# Central Tendency of price, country and variety

```python
df.describe()
```

```python
uniq_countries = df.country.unique()
len(uniq_countries)
```

```python
value_count = df.country.value_counts()
value_count
```

```python
df.price.value_counts()
```

```python
best_price = df.groupby('points')["price"].min()
best_price
```

```python
max_price_each_variety = df.groupby("variety")["price"].max()
max_price_each_variety
```

```python
min_price_each_variety = df.groupby("variety")["price"].min()
min_price_each_variety
```

```python
max_min_price_each_variety = df.groupby("variety")["price"].agg(["min", "max"])
print(max_min_price_each_variety)
```

```python
# Most expensive wine variety

sorted_expensive_variety = max_min_price_each_variety.sort_values(by= ["max"], ascending = False)
sorted_expensive_variety
```

```python
# Most common pair of Two variables (Country, Wine variety)

gr = df.groupby(["country", "variety"])
gen = gr.size().sort_values(ascending = False)
gen
```

```python
# Most common selling wine in india

ind = df[df["country"] == "India"]
ind["variety"].value_counts()
```
