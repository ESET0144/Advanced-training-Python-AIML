## Pandas

```python
def rows_in_time_range(df, time_column, start_ts_str, timedelta_str):
  # Returns roe from df, where start_ts < time_column <= start_ts + delta
  start_ts = pd.to_datetime(start_ts_str).tz_localize("US/Pacific")
  # start_ts = pd.to_datetime(start_ts_str) # invalid timezone comparison
  end_ts = start_ts + pd.Timedelta(timedelta_str)
  #return df.query(f"{time_column} >= '{start_ts}' and {time_column} <= '{end_ts}'")
  return df.query("@start_ts <= {} < @end_ts".format(time_column))
```

```python
rows_in_time_range(df, 'time', '2022-09-14 00:00', '52 minutes 31 seconds')
```
